DRAW A LINE 

```python
data = image.imread('/content/Screenshot (153).png')
x=[100,200]
y=[400,200]
plt.plot(x,y,linewidth ='3',color = 'black')
plt.imshow(data)
plt.show()
```

DRAW TWO LINES (INTERSECTING LINES ALSO---BASED ON THE COORDINATES)

```python
data = image.imread('/content/Screenshot (153).png')
x1=[100,200]
y1=[400,200]
x2=[300,400]
y2=[500,300]
plt.plot(x1,x2,y1,y,linewidth ='3',color = 'black')
plt.imshow(data)
plt.axis('off')
plt.show()
```

DRAW A RECTANGLE 

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches 
from PIL import Image 
import numpy as np 
```

```python
data = np.array(Image.open('/content/Screenshot (153).png'),dtype=np.uint8)
plt.imshow(data)
fig,ax=plt.subplots(1)
ax.imshow(data)
rect=patches.Rectangle((125,250),200,400,linewidth='3',edgecolor='pink',facecolor='blue')
ax.add_patch(rect)
plt.show()
```
